# Pivot Tables

Pivot tables take column-wise data as input & group the entries into a two-dimensional summary table.

Let's use the database of passengers on the *Titanic*, available through the Seaborn library.

````python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
````

````python
titanic.head()
````

Start by grouping according to gender, survival status, or some combination thereof. You might be tempted to apply a ``GroupBy`` operation.

````python
titanic.groupby('sex')[['survived']].mean()
````

If we look at survival by both sex & class, we might: *group by* class and gender, *select* survival, *apply* a mean aggregate, *combine* the resulting groups, and then *unstack* the hierarchical index:

````python
titanic.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack()
````

Two-dimensional ``GroupBy`` operations are common enough that Pandas includes a convenience routine, ``pivot_table``.

````python
titanic.pivot_table('survived', index='sex', columns='class')
````

### Multi-level pivot tables

Pivot tables can be specified with multiple levels, and via a number of options. For example, we might be interested in looking at age as a third dimension. We'll bin the age using the __[``pd.cut``](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.cut.html)__ function:

````python
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class')
````

Repeat the strategy when working with columns; let's add info on the fare paid using __[``pd.qcut``](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.qcut.html)__ to automatically compute quantiles:

````python
fare = pd.qcut(titanic['fare'], 2)
titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
````

### Additional pivot table options

Here's the ``pivot_table`` options:

```python
# call signature as of Pandas 0.18
DataFrame.pivot_table(data, values=None, index=None, columns=None,
                      aggfunc='mean', fill_value=None, margins=False,
                      dropna=True, margins_name='All')
```

``fill_value`` and ``dropna``, handle missing data.

__``aggfunc`` controls the aggregation method__ (mean = default). It can be a string (e.g., ``'sum'``, ``'mean'``, ``'count'``, ``'min'``, ``'max'``, etc.) or a function (e.g., ``np.sum()``, ``min()``, ``sum()``, etc.). It can also be specified as a dictionary mapping a column to any of the above desired options.

````python
titanic.pivot_table(index='sex', columns='class', aggfunc={'survived':sum, 'fare':'mean'})
````

Note the lack of a ``values`` keyword; when specifying a mapping for ``aggfunc``, this is determined automatically.

Find the totals for each grouping with ``margins``:

````python
titanic.pivot_table('survived', index='sex', columns='class', margins=True)
````

This displays class-agnostic survival rate by gender, gender-agnostic survival rate by class, and an overall survival rate of 38%. The margin label can be specified with the ``margins_name`` keyword, which defaults to ``"All"``.

### Example: Birthrate Data

Download US birth data from the Centers for Disease Control (CDC) at https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv.

````python
!curl -O https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv
````

````python
births = pd.read_csv('births.csv')
births.head()
````

Add a decade column. Review male and female births as a function of decade.

````python
births['decade'] = 10 * (births['year'] // 10)
births.pivot_table('births', index='decade', columns='gender', aggfunc='sum')
````

Plot the total number of births by year.

````python
%matplotlib inline
import matplotlib.pyplot as plt
sns.set()  # use Seaborn styles
births.pivot_table('births', index='year', columns='gender', aggfunc='sum').plot()
plt.ylabel('total births per year');
````

Over the past 50 years male births have outnumbered female births by around 5%.

### Further data exploration

Start by cleaning the data: removing outliers caused by mistyped dates (e.g., June 31st) or missing values (e.g., June 99th). __One easy way to remove these all at once is to cut outliers; we'll do this via a sigma-clipping operation.__

````python
quartiles = np.percentile(births['births'], [25, 50, 75])
mu = quartiles[1]
sig = 0.74 * (quartiles[2] - quartiles[0])
````

The final line is an estimate of the sample mean, where the 0.74 comes from the interquartile range of a Gaussian distribution. Now use ``query()`` to filter out rows with births outside these values:

````python
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')
````

Set the ``day`` column to integers; previously it had been a string because some columns in the dataset contained the value ``'null'``:

````python
births['day'] = births['day'].astype(int)
````

Combine the day, month, and year to create a Date index. This allows us to quickly compute the weekday corresponding to each row:

````python
births.index = pd.to_datetime(
    10000 * births.year + 
    100 * births.month + 
    births.day, 
    format='%Y%m%d')

births['dayofweek'] = births.index.dayofweek
````

Plot births by weekday for several decades:

````python
import matplotlib.pyplot as plt
import matplotlib as mpl

births.pivot_table('births', index='dayofweek',
                    columns='decade', aggfunc='mean').plot()
plt.gca().set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.ylabel('mean births by day');
````

Apparently births are slightly less common on weekends. Note the 1990s and 2000s are missing because the CDC data contains only the month of birth starting in 1989.

Plot the mean number of births by the day of the *year*. First group the data by month and day separately:

````python
births_by_date = births.pivot_table('births', 
                                    [births.index.month, births.index.day])
births_by_date.head()
````

The result is a multi-index over months and days. Convert the indexes into a date by associating them with a dummy year variable (making sure to choose a leap year so February 29th is correctly handled!)

````python
births_by_date.index = [pd.datetime(2012, month, day)
                        for (month, day) in births_by_date.index]
births_by_date.head()
````

We now have a time series reflecting the average number of births by date of the year. ``plot`` method to plot the data. It reveals some interesting trends:

````python
fig, ax = plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=ax);
````
